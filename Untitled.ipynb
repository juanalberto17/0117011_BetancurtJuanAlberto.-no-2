```python
# ejercicio 12
abc = 'abcdefghijklmnñopqrstuvwxyz'
reemplazar = (('á','a'),
              ('é','e'),
              ('í','i'),
              ('ó','o'),
              ('ú','u'),
              ('ü','u'))
word1 = str(input("ingrese la primera palabra:"))
word2 = str(input("ingrese la segunda palabra:"))
w1 = word1
w2 = word2


word1 = word1.lower()
word2 = word2.lower()
while word1 == word2:
    print("no pueden ser las mismas palabras, intente otra vez:")
    word1 = str(input("ingrese la primer palabra:"))
    word2 = str(input("ingrese la segunda palabra:"))


for a,b in reemplazar:
    word1 = word1.replace(a, b)
    word2 = word2.replace(a, b)


band = 3
i = 0
l1 = -1
l2 = -1
while band != 1:            
    if l1 != l2:
        band = 1
    else:    
        if i >= len(word1):
            band = 1
            l1 = 2
            l2 = 3
        if i >= len(word2):
            band = 1
            l1 = 2
            l2 = 1
            
    while l1 < 0 or l2 < 0:
        for j in range(len(abc)):
            if word1[i] == abc[j]:
                l1 = j
            if word2[i] == abc[j]:
                l2 = j
    i +=1

    
if l1 < l2:
    print("Primera palabra:", w1)
    print("Segunda palabra:", w2)
elif l2 < l1:
    print("Primera palabra:", w2)
    print("Segunda palabra:", w1)
else:
    print("las dos palabras son iguales.") 

```
